```python
# Expanding contractions
from contractions import CONTRACTION_MAP
def expand_contractions(text, contraction_mapping=CONTRACTION_MAP):
    
    contractions_pattern = re.compile('({})'.format('|'.join(contraction_mapping.keys())), 
                                      flags=re.IGNORECASE|re.DOTALL)
    def expand_match(contraction):
        match = contraction.group(0)
        first_char = match[0]
        expanded_contraction = contraction_mapping.get(match)\
                                if contraction_mapping.get(match)\
                                else contraction_mapping.get(match.lower())                       
        expanded_contraction = first_char+expanded_contraction[1:]
        return expanded_contraction
        
    expanded_text = contractions_pattern.sub(expand_match, text)
    expanded_text = re.sub("'", "", expanded_text)
    return expanded_text
test = expand_contractions(test)
test
```

```python
# Removing HTML Tags such as &amp that could be seen in above data
import re

def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
test = cleanhtml(test)
test
```

```python
# Trying to convert numbers to words
import inflect
p = inflect.engine()
p.number_to_words(test)
```

```python
# Removing special characters and punctuations
def remove_special_characters(text, remove_digits=False):
    pattern = r'[^a-zA-z0-9\s]' if not remove_digits else r'[^a-zA-z\s]'
    text = re.sub(pattern, '', text)
    return text

special_char_pattern = re.compile(r'([{.(-)!}])')
test = special_char_pattern.sub(" \\1 ", test)
test = remove_special_characters(test)
test
```

```python
test = re.sub(' +', ' ', test)
test
```

```python
nlp = spacy.load('en_core_web_sm')
def lemmatize_text(text):
    text = nlp(text)
    text = ' '.join([word.lemma_ if word.lemma_ != '-PRON-' else word.text for word in text])
    return text

test = lemmatize_text(test)
test
```

```python
# Removing stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english')) 

def remove_stopwords(text):
    tokens = word_tokenize(text)
    tokens = [token.strip() for token in tokens]
    filtered_tokens = [token for token in tokens if token not in stop_words]
        
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text

test = remove_stopwords(test)
test
```

```python
def remove_punctuations(text):
    text = re.sub(r'[^a-zA-z\s]', '', text)
    return text

special_char_pattern = re.compile(r'([{.(-)!}])')
test = special_char_pattern.sub(" \\1 ", test)
test = remove_punctuations(test) 
test
```

```python
# Remove punctuations and special characters
test = re.sub(' +', ' ', test)
test
```

```python
# Complete function for text preprocessing
stop_words = set(stopwords.words('english')) 

def text_preprocessing(text):
    # Lowercasing the text
    text = text.lower()
    
    # Removing html data
    text = cleanhtml(text)
    
    # expanding contractions
    text = expand_contractions(text)
    
    # lemmatize text
    text = lemmatize_text(text)
    
    # removing punctuations
    special_char_pattern = re.compile(r'([{.(-)!}])')
    text = special_char_pattern.sub(" \\1 ", text)
    text = remove_punctuations(text)
    
    # removing extra whitespace
    text = re.sub(' +', ' ', text)
    
    # removing stopwords
    text = remove_stopwords(text)
    
    # stripping extra space
    text = text.strip()
    
    return text
```
